```python
movies = pd.read_csv('D:\SEM 2\Python for analytics\jupyter\ml-latest-small\ml-latest-small\\movies.csv', )
movies.head()
```

```python
movies.rename(columns = {'title':'old_title'}, inplace=True)
movies.head()
```

```python
movies.sort_values("old_title", inplace = True) 
movies.drop_duplicates(subset ="old_title", keep = 'first', inplace = True)
movies.head()
```

```python
movies['release_year'] = movies['old_title'].str[-5:-1]
movies['title'] = movies['old_title'].str[:-6]
movies.head()
```

```python
#unique values
movies.release_year.unique()

#count of movieId and release year
movies_grouped = movies.pivot_table(['movieId'], 
                                    index=['release_year'], aggfunc='count')
#movies_grouped
movies_grouped.sort_values('movieId',ascending=False)

#subset for unique values array 'not_an_year'
not_an_year = ['Bab', 'Roa', '008', '012', '014', '973', '995', '998', '999', 'atso', 'erso', 'he O', 'imal', 'irro','ligh', 'lon', 'osm']

#convert the subset array into a dataframe 'remove_year'
remove_year = pd.DataFrame(not_an_year)
remove_year.rename(columns = {0:'year_flag'}, inplace=True)
remove_year   

#merge remove file to the 'movies' file to create a flag 'year_flag'
movies = pd.merge(left=movies, right=remove_year, how='left', left_on='release_year', right_on='year_flag')
movies.head()

```

```python
movies = movies.drop(['old_title', 'year_flag'], axis = 1) 
movies.head()
```

```python
movies_base_file = pd.merge(left=movies_base_file, right=movies, how='left', on='movieId')
movies_base_file.head()
```
